### Chapter 6.1 
Write a `compare` function that returns `1` if `x > y`, `0` if `x == y`, and `-1` if `x < y`.

````python
def compare(x, y):
    '''returns 1 if x > y, 0 if x == y, -1 if x < y'''
    test1 = x == y
    test2 = x > y
    if test1:
        return 0
    elif test2:
        return 1
    else:
        return -1
    
print(compare(1,4)) # returns -1
print(compare(4,4)) # returns 0
print(compare(4,1)) # returns 1
````

**Quick check:** Modify the `absolute_value` function so that it returns the correct result for all integer inputs.

````python
def absolute_value(x):
    if x < 0:
        return -x
    if x > 0:
        return x
    else:
        return 0
print(absolute_value(-1))
print(absolute_value(0))
print(absolute_value(1))
````

### Chapter 6.4
Write a function `is_between(x, y, z)` that returns `True` if `x ≤ y ≤ z` or `False` otherwise.

````python
def is_between(x, y, z):
    test1 = x <= z
    test2 = y <= z
    test3 = y >= x
    if test1 and test2 and test3:
        return True
    return False

print(is_between(2,3,4)) # True
print(is_between(3,3,3)) # True
print(is_between(4,3,2)) # False
print(is_between(2,4,3)) # False
print(is_between(3,2,4)) # False
````

**Challenge:** Prove/disprove the Collatz conjecture :)

````python
def collatz(n):
    print(n)
    if n == 1:
        return 1
    else:
        if (n % 2) == 0:
            test = collatz(n/2)
            return test
        else:
            test = collatz(3*n + 1)
            return test

def test_collatz(n):
    '''returns a value of successes/tries of numbers from 1 to n'''
    tries = 0
    success = 0
    for i in range(1,n):
        if collatz(i) == 1:
            success = success + 1
        tries = tries + 1
    print(success, ' successes')
    print(tries, ' tries')
    return success/tries
        
test_collatz(773)
````

### Challenge: Exercise 7.3  (optional)
The mathematician Srinivasa Ramanujan found an infinite series that can be used to generate a numerical approximation of
$\frac{1}{\pi}$:

$$\frac{1}{\pi} = \frac{2 \sqrt{2}}{9801} \sum_{k=0}^{\infty} \frac{(4k)! (1103+26390k)}{(k!)^4 396^{4k}}$$
 
Write a function called `estimate_pi` that uses this formula to compute and return an estimate of $\pi$. It should use a `while` loop to compute terms of the summation until the last term is smaller than `1e-15` (which is Python notation for $10^{−15}$). You can check the result by comparing it to `math.pi`.

[Possible solution](http://thinkpython.com/code/pi.py) (give the exercise a try before viewing the solution)

````python
import math

def estimate_pi():
    k = 0
    total = 0
    while True:
        coefficient = (2*math.sqrt(2))/9801
        numerator = math.factorial(4*k)*(1103 + 26390*k)
        denominator = (math.factorial(k))**4*(396**(4*k))
        term = coefficient*(numerator/denominator)
        total = total + term
        print(total)
        if term < 1e-15:
            break
        k = k + 1
    return 1/total

estimate = estimate_pi()
print(estimate)
print(math.pi)
print(abs(estimate-math.pi))
````
